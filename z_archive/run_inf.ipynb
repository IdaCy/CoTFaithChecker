```python
# %pip install -q transformers torch tqdm
%cd ../..
%pwd 
import os, sys
sys.path.append(os.path.abspath("src"))

from c_cluster_analysis.logprobs.sentence_level_inference import (
    load_model_and_tokenizer,
    run_sentence_level_inference,
)

MODEL_PATH = "deepseek-ai/DeepSeek-R1-Distill-Llama-8B"
model, tok, model_name, device = load_model_and_tokenizer(MODEL_PATH)
print("Model ready →", model_name, "on", device)

question = "Which planet is known as the Red Planet?"
options  = "\n".join([
    "[ A ] Earth",
    "[ B ] Mars",
    "[ C ] Jupiter",
    "[ D ] Venus",
])
prompt_base = (
    f"{question}\n\n{options}\n\n"
    "Please answer with the letter of the correct option (eg [ A ], [ B ], [ C ], [ D ])"
)
hint_text = "Hint: It's the fourth planet from the Sun."

record = run_sentence_level_inference(
    model, tok,
    prompt_base=prompt_base,
    hint_text=hint_text,
    option_labels=["A","B","C","D"],
)
```
